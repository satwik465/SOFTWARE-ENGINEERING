```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
  temperature = a*time**2 + b*time + c
  return temperature

def main():
  # Time values
  time_values = np.linspace(0, 10, 50)

  # Hard-coded coefficients sets
  coefficient_sets = [
      (1, 1, 30),
      (2, 1, 20),
      (1, 2, 10)
  ]

  for i, (a, b, c) in enumerate(coefficient_sets):
    temperature = quadratic_model(time_values, a, b, c)
    plt.plot(time_values, temperature, label=f'Set {i+1} a=1 , b=1, c=30')

  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modelling with Quadratic Equation (Multiple Hard-coded Sets)')
  plt.show()

if __name__ == '__main__':
  main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * time**2 + b * time + c
    return temperature

file_paths = ['file1.txt', 'file2.txt', 'file3.txt']  # Replace with your file paths

time_values = np.linspace(0, 50, 51)

for file_path in file_paths:
    try:
        with open(file_path, 'r') as file:
            values = file.readline().strip().split()
            if len(values) == 3:
                a = float(values[0])
                b = float(values[1])
                c = float(values[2])
                label = f"{file_path} (a={a}, b={b}, c={c})" # Added label formatting
            else:
                raise ValueError("File should contain exactly 3 values.")

        temperatures = quadratic_temperature_model(time_values, a, b, c)
        plt.plot(time_values, temperatures, label=label) # Use formatted label

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except ValueError as e:
        print(f"Error: {e}")

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Weather Modelling with quadratic equation')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def read_coefficients(filename):
    with open(filename, 'r') as f:
        # Split the line into individual numbers and convert each to float
        coefficients = [float(num) for num in f.read().split()]
    return coefficients

def quadratic_model(time, coefficients):
    a, b, c = coefficients
    temperature = a*(time**2) + b*time + c
    return temperature

def main():
    coefficients = read_coefficients('1743.txt')
    time_values = np.linspace(0, 51, 50)
    temperature_values = quadratic_model(time_values, coefficients)
    plt.plot(time_values, temperature_values, label='Coefficients from file   a=15, b=19, c=33')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather modeling with quadratic equation (coefficientsfrom file)')
    plt.show()

if __name__ == "__main__":
    main()
```
